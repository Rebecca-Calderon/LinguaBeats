````python
import json

# Making a function to add a new song to the JSON data
def create_song(json_data, song_name, artist, album, duration, wps, originalLyrics, translatedLyrics, quizzes):
    new_song = {
        "songID": None,
        "songName": song_name,
        "artist": artist,
        "album": album,
        "duration": duration,
        "wps": wps,
        "originalLyrics": originalLyrics,
        "translatedLyrics": translatedLyrics,
        "Quiz": quizzes
    }
    # Here we are appending the new song object to the existing array
    return new_song

# Loading existing JSON data from the file, checking it it exists, if it doesn't we create a new structure
def load_existing():
    data = None
    try:
        with open("songs.json", "r") as json_file:
            data = json.load(json_file)
    except FileNotFoundError:
        data = {
            "songs": {}
        }
    return data
def save_json(data):
    # Save the updated JSON data to the file
    with open("songs.json", "a") as json_file:
        json.dump(data, json_file, indent=4)
````

````python
from bs4 import BeautifulSoup
import urllib.request
def linkbuilder(artist, name):
        link = "https://lyricstranslate.com/en/songs/0/" + artist + "/" + name
        return link
def getLyricTranslate(artist, name):
    html = urllib.request.urlopen(linkbuilder(artist,name)).read()
    soup = BeautifulSoup(html, 'html.parser')
    stt = soup.find(class_="stt")
    a = stt.find('a', href=True)
    songlink= a.get("href")

    spanish = urllib.request.urlopen("https://lyricstranslate.com/"+songlink).read()
    spanishsoup = BeautifulSoup(spanish, 'html.parser')
    spanishtext = spanishsoup.find(id="song-body")
    spanishLyrics = spanishtext.get_text().split("\n")

    for s in spanishLyrics:
        if("[" in s and "]" in s):
            spanishLyrics.remove(s)
    language = spanishsoup.find(id="song-lang-selector")
    engID = ''
    for lang in language.find_all('option'):
        if(lang.text == 'English'):
            engID = lang['value']
    english_link = "https://lyricstranslate.com/en/node/"+engID
    english = urllib.request.urlopen(english_link).read()
    englishsoup = BeautifulSoup(english, 'html.parser')
    englishtext = englishsoup.find(id="song-body")
    englishLyrics = englishtext.get_text().split("\n")
    for e in englishLyrics:
        if("[" in e and "]" in e):
            englishLyrics.remove(e)
    return spanishLyrics, englishLyrics
def getLyrics(artist, name):
    artist = artist.replace(" ", "%2520")
    name = name.replace(" ", "%2520")
    originalLyrics, translatedLyrics = getLyricTranslate(artist, name)
    return originalLyrics, translatedLyrics
import requests
import json
# Set up OpenAI API credentials
API_KEY = 'api-key'
API_ENDPOINT = "https://api.openai.com/v1/chat/completions"

def generate_chat_completion(messages, model="gpt-4", temperature=1, max_tokens=500):
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {API_KEY}",
    }

    data = {
        "model": model,
        "messages": messages,
        "temperature": temperature,
    }

    if max_tokens is not None:
        data["max_tokens"] = max_tokens

    response = requests.post(API_ENDPOINT, headers=headers, data=json.dumps(data))

    if response.status_code == 200:
        return response.json()["choices"][0]["message"]["content"]
    else:
        raise Exception(f"Error {response.status_code}: {response.text}")
def createQuiz(difficulty, originalLyrics, translatedLyrics):   
    system="You are a Spanish teacher that uses lyrics to teach Spanish. Given song lyrics in Spanish and English and a difficulty level (1-5), create the following quizzes each quiz will have 5 questions each. Dont use the same vocabulary between quizzes. If it helps, make a list of vocabulary words in advance and then assign them to quizzes: Quiz 1 (Basic Match Word to Meaning): the format will be \"What does ____ mean?\" followed by multiple-choice  in bullet format or free response (at least 2 free response) Quiz 2 (Fill in the blank): The format will be a line from the Spanish lyrics with a vocabulary word blanked out, followed by the English translation of that line (source from the English lyrics). You should prioritize verbs, nouns, and adjectives rather than filler ‘stop’ words Quiz 3 (Match word to audio): For this each question is just a vocabulary word from the song that is hard to pronounce Quiz 4 (Match word to picture): For this each question is just a vocabulary word from the song that is easy to symbolize with a picture i.e respond with \"Noche -- picture of a night sky\" FOR ALL VOCABULARY, AVOID STOP WORDS, GRAMMER WORDS, NAMES OF BRANDS, AND UNCOMMON WORDS"
    prompt="Difficulty level " + str(difficulty) + "Spanish lyrics"+"      ".join(spanishLyrics)+"               English Lyrics"+"      ".join(englishLyrics)
    messages = [
        {"role": "system", "content": system},
        {"role": "user", "content": prompt}
    ]
    response_text = generate_chat_completion(messages)
    return response_text
def getSong(song):
    songID = song["artist"][0] + song["album"][0] + song["name"]
    #first we check to see if the song is already in the cache
    data = load_existing()
    if(songID not in data):
        originalLyrics, translatedLyrics = getLyrics(song["artist"],song["name"])
        wps = len(" ".join(originalLyrics))/song["duration"]
        quizzes =  {1: createQuiz(1, originalLyrics, translatedLyrics), 
                    2: createQuiz(2, originalLyrics, translatedLyrics), 
                    3: createQuiz(3, originalLyrics, translatedLyrics), 
                    4: createQuiz(4, originalLyrics, translatedLyrics), 
                    5: createQuiz(5, originalLyrics, translatedLyrics)}
        new_song = create_song(data, song["name"], song["artist"], song["album"], song["duration"], wps, originalLyrics,translatedLyrics, quizzes)       
        new_song["songID"] = songID
        save_json(new_song)
        data[new_song["songID"]] = new_song
    else:
        return data        
thisSong = getLyrics(current_song_metadata["artist"], current_song_metadata["name"] )
````
